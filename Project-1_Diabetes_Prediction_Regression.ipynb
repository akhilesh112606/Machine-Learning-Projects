```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data
# Here in this step we will be loading the data from scikit learn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
# Problem Statement : 
# To Predict the disease progression one year after baseline
# with provided 10 features
```

```python
print(diabetes.DESCR)
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
data.head()
```

```python
data['target'] = diabetes.target
data['target']
```

```python
data
```

```python
# EDA, DATA CLEANING, DATA PREPARATION, FEATURE ENGINEERING has to be done
# But there all this are already been done.
# It is complete filtered and clean dataset
```

```python
x = data[['bmi']] # Independent Variable
y = data['target'] # Dependent Variable
```

```python
print(x, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 1)
```

```python
# Randome State helps us to reproduce the same elements or splitting
# Again and again with same order
```

```python
# Model Training!
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
# Training of the Model
model.fit(x_train, y_train)
# At the backend it will be using Gradient Descent to give optimal Coeff
```

```python
model.coef_ # Attribute of the Model
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test) # We have Predicted the X-Test Values
```

```python
# Visualizing the result
plt.scatter(x_test, y_test, color = 'blue', label = 'Actual')
plt.plot(x_test, y_pred, color = 'black', linewidth = 3, label = 'Predicted Linear Regression')
plt.xlabel("BMI")
plt.ylabel("Diabetes Level")
plt.title("Linear Regression on Diabetes Data")
plt.legend()
plt.show()
```

```python
# Model Evalutaion
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
